# **Definition of variables**



1.   UNITID: Its an unique ID for the universities and can act a primary key.
2.   INSTNM: Name of the Institutions.
3.   CITY: City in which the University is located.
4.   STABBR: It denotes the State Code of the american states where the university is located.
5.   ZIP: Its the zipcode of the location where the university is located.
6.   ADM_RATE: It is the acceptance rate of the university also known as admission  rate.
7.   UGDS: The Utrecht Gender Dysphoria Scale (UGDS) is a validated, 12-item screening measure for both adults and adolescents used extensively in gender clinics to assess gender dysphoria. [Source](https://https://www.ncbi.nlm.nih.gov/pmc/articles/PMC7430422/)
8.   TUITIONFEE_IN: It is the tuition fee of the university.
9.   MN_EARN_WNE_P10: This translates to the mean earnings of graduates working 10 years after entry. [Source](https://www.zippia.com/advice/junior-colleges-with-highest-earning-graduates-in-each-state/#:~:text=In%20data%2Dmining%20talk%2C%20the,accurately%20rank%20the%20junior%20colleges.)

**Calculating the min, max, and average in-state tution of the above four institutions.**

```python
df2= df2[(df2['INSTNM'] =='Johns Hopkins University') | (df2['INSTNM'] =='Loyola University Maryland') | (df2['INSTNM'] =='University of Maryland-College Park') | (df2['INSTNM'] =='University of Maryland-Baltimore County')] #selecting the rows with following univeristy name to find the values asked
minimum_value =df2['TUITIONFEE_IN'].min() #finding the minimum value
print("minimum_value:" ,minimum_value)

maximum_value =df2['TUITIONFEE_IN'].max() #finding the maximum value
print("maximum_value:" ,maximum_value)

Avg_value =df2['TUITIONFEE_IN'].mean() #finding the average value
print("average_value:" ,Avg_value)
```

**smaller dataset further to only keep colleges from Maryland, Virgina, Washington DC, Delaware, and Pennsylvinia.**

```python
df3 = df2.loc[(df2['STABBR'] =='MD') | (df2['STABBR'] =='PA') | (df2['STABBR'] == 'DC') | (df2['STABBR'] == 'VA') | (df2['STABBR'] == 'DE')] #creating a new dataframe with states Maryland, Virgina, Washington DC, Delaware, and Pennsylvinia.
df3
```
