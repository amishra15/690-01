```python
import pandas as pd
from datetime import datetime

def filter_data(b_string):
  return int(b_string.replace('2010*', '2010'))

df['Census_Year'] = df['Census Year'].apply(filter_data)
```

```python

df['Census Cost']= df['Census Cost'].replace('$13 Billion', '1300000000')
```

```python
df['Census Cost']= df['Census Cost'].replace('$4.5 Billion', '4500000000')
```

```python
def filter_data(b_string):
  return int(b_string.replace(',', '').replace('$', ''))

df['Census Cost1'] = df['Census Cost'].apply(filter_data)
df
```

```python
def filter_data(b_string):
  return (b_string.replace('$', '').replace('cents', ''))

df['Average Cost Per Person1'] = df['Average Cost Per Person'].apply(filter_data).astype(float)
df['Average Cost Per Person1'].loc[18:23] = round(df['Average Cost Per Person1'].loc[18:23]*100,2)
df
```

```python
df = df[['Census_Year', 'Tot Pop', 'Census Cost1', 'Average Cost Per Person1']]
df
```

```python
df.describe(include="all")
```

```python
import plotly.express as px
```

```python
fig = px.bar(
    data_frame=df,
    title="average cost per person over time",
    orientation="v",
    x="Census_Year",
    y="Average Cost Per Person1"
)

fig.update_xaxes(tickangle=45)

fig.show()
```

```python
fig = px.scatter(
    data_frame=df,
    title="Total Population vs Total Cost",
    orientation="v",
    x="Tot Pop",
    y="Census Cost1",
    template = 'plotly_dark'
)

fig.update_xaxes(tickangle=45)

fig.show()
```
